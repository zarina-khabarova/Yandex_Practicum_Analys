### Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
#объявим запрос
query = '''
            SELECT b.book_id,
                   b.title,
                   COUNT (DISTINCT re.review_id) AS count_review,
                   ROUND(AVG(rating),2) AS avg_rating
            FROM books AS b
            LEFT JOIN reviews AS re ON b.book_id = re.book_id
            LEFT JOIN ratings AS ra ON b.book_id = ra.book_id
            
            GROUP BY b.book_id, b.title
            ORDER BY count_review DESC
            
'''
```

```python
#выведем полученный запрос
pd.io.sql.read_sql(query, con = engine)
```

**Вывод:**

<p></p>

<div style="border:solid green 2px; padding: 20px">
Общее количество книг и брошюр в катологе составляет 1000 штук. Количество обзоров на книгу находится в диапазоне от 0 до 7. То есть в датасете присутствуют книги, на которые нет обзоров. Так, лидер по количеству обзоров книга "Twilight (Twilight #1)", количество обзоров 7, средний рейтинг 3,66. Если обратить внимание на средную оценку, то можно увидеть, что количество обзоров никак не связано со средней оценкой, то есть корреляция отсутствует. Есть книга, например, "The Cat in the Hat and Other Dr. Seuss Favorites", у которой нет обзоров, при этом средний рейтинг = 5. 

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
#объявим запрос
query = '''
            WITH
            t_1 AS
            (SELECT publisher_id,
                    book_id
            FROM books
            WHERE num_pages > 50)
            SELECT t_1.publisher_id,
                   p.publisher
            FROM t_1
            INNER JOIN publishers AS p ON t_1.publisher_id = p.publisher_id
            GROUP BY t_1.publisher_id, p.publisher 
            ORDER BY COUNT(t_1.book_id) DESC
            LIMIT 1
'''
```

```python
#выведем полученный запрос
pd.io.sql.read_sql(query, con = engine)
```

**Вывод:**

<p></p>

<div style="border:solid green 2px; padding: 20px">
По условию задачи были исключены из анализа книги, в которых меньше 50 страниц, так как они относятся к брошюрам. Издательство, которое выпустило больше всего книг толще 50 страниц - Penguin Books (британское издательство). 

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
#объявим запрос
query = '''
            WITH 
            t1 AS
            (SELECT book_id
            FROM ratings
            GROUP BY book_id
            HAVING COUNT(rating_id) >= 50)
            SELECT a.author_id, 
                   a.author, 
                   AVG(r.rating) AS avg_rating
            FROM books AS b

            INNER JOIN t1 ON b.book_id = t1.book_id
            INNER JOIN authors AS a ON b.author_id = a.author_id
            INNER JOIN ratings AS r ON t1.book_id = r.book_id
            GROUP BY a.author_id, a.author
            ORDER BY AVG(r.rating) DESC
            LIMIT 1

'''
```

```python
#выведем полученный запрос
pd.io.sql.read_sql(query, con = engine)
```

**Вывод:**

<p></p>

<div style="border:solid green 2px; padding: 20px">
По условию задачи были исключены книги из анализа, в которых меньше 50 оценок. Автором с самой высокой средней оценкой книг является J.K. Rowling/Mary GrandPré. Средняя оценка составила 4,28. Джоан Роулинг самая популярная британская писательница, написавшая серию книг о Гарри Поттере. 

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
#объявим запрос
query = '''
           SELECT COUNT(DISTINCT rv.text)/ COUNT(DISTINCT rv.username) AS avg_rv_count
           FROM ratings AS r

           INNER JOIN reviews AS rv ON r.username = rv.username
           WHERE r.username in  
                    (SELECT username
                    FROM ratings
                    GROUP BY username
                    HAVING COUNT(rating_id) > 50)

'''
```

```python
#выведем полученный запрос
pd.io.sql.read_sql(query, con = engine)
```

Пользователи, которые поставили больше 50 оценок, написали в среднем обзоры на 24 книги и брошюры.
